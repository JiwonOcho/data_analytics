### 8개의 샘플 파일 데이터 행 추출
1) KL_BOOK_TBL_202301271758.csv 

```python
book = pd.read_csv('KL_BOOK_TBL_202301271758.csv', encoding = 'cp949')
```

### 데이터 확인

```python

book.info
```

```python
book_columns = book.columns
print(book_columns)
print('\nbook_columns의 개수는',len(book_columns))
```

### 결측치 확인

```python
book.isnull()
```

### 결측치 개수 확인

```python
nlist=[]
nlist.append(book.isnull().sum())
print(nlist)
```

```python
for i in book_columns:
    print(book[i][book[i].isna()])
```

```python
na_list = []
for i in book_columns:
    na_list.append(book[i][book[i].isna()])
print(na_list.index('Name'))
```

```python
#columns안에서 어느 행에 있는지 파악하기
for list in na_list:
    if len(list) != 0:
        print(list, '\n결측치 개수는',len(list))
```

```python
print(na_list)
```

```python
#결측치 위치 파악
book.loc[book['VOL_TITLE'].isna()].head()
```
