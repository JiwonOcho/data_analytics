### 금액별 구매건수 알아보기

```python
payment = [] # 각 금액 값을 담을 리스트 payment 선언
bigger = [] # 만원 초과인 값들을 담아둘 리스트 bigger 선언
smaller = [] # 만원 이하인 값들을 담아둘 리스트 smaller 선언


# 카드 금액 모두 payment 리스트에 담기
for row in data:
    payment.append(int(row[6]))
#print(f'총 금액: {payment}')

# 조건에 맞게 리스트에 담기 
for i in payment:
    if i > 10000:
        bigger.append(i)
    elif i <= 10000:
        smaller.append(i)
        
print(f'만원 이하 금액 사례들:{smaller}')
print(f'만원 초과 금액 사례들:{bigger}')
print(f'모든 금액 사례들:{payment}')
```

```python
print(f'총 만원 이하 구매 건수는 {len(smaller)}건 입니다.')
print(f'총 만원 초과 구매 건수는 {len(bigger)}건 입니다.')
print(f'총 구매 건수는 {len(payment)}건 입니다.')
```

#### 9.2.4 지출액 상위 10개 항목 뽑기

```python
spending = {}

for row in data : 
    if row[-1] == '전표매입' : 
        store, payment = row[-4], int(row[-3])
        if store not in spending.keys() : 
            spending[store] = payment
        else : 
            spending[store] += payment

print(spending)
```

```python
import operator
import matplotlib.pyplot as plt

top10 = sorted(spending.items(), key = operator.itemgetter(1), reverse=True)[:10]

top10_store = []
top10_amount = []

for t in top10 : 
    top10_store.append(t[0])
    top10_amount.append(t[1])

print(top10_store)
print(top10_amount)

plt.rc('font', family = 'NanumGothic')
plt.title('10-12월 지출 TOP 10')
plt.barh(top10_store, top10_amount, color = 'b')
plt.ticklabel_format(axis='x', style= 'plain')
plt.show()
```
